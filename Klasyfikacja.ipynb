```python
import pandas as pd

data = pd.read_csv('Dane/dane_treningowe.csv', sep=';')
```

```python
data
```

```python
data = data.set_index('ID_KLIENTA')
```

```python
data = data[data['RODZAJ_UBEZPIECZENIA'] != 0]
```

```python
y = data['RODZAJ_UBEZPIECZENIA']
```

```python
x = data.drop(columns=['RODZAJ_UBEZPIECZENIA'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
kneib = KNeighborsClassifier(n_neighbors=5)
kneib.fit(X_train, y_train)
accuracy_score(y_test, kneib.predict(X_test))
```

```python
import random
```

```python
accuracy_score(y_test, random.choices(kneib.classes_, k=len(y_test)))
```
