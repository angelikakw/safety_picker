```python
import pandas as pd

data = pd.read_csv('dane_treningowe_OH.csv', sep=';')
```

```python
data
```

```python
data = data.set_index('ID_KLIENTA')
```

```python
data.columns
```

```python
polisa_columns = ['polisa__1.0', 'polisa__2.0', 'polisa__3.0',
       'polisa__5.0', 'polisa__6.0', 'polisa__7.0', 'polisa__8.0',
       'polisa__9.0', 'polisa__10.0', 'polisa__13.0', 'polisa__14.0',
       'polisa__15.0']
```

```python
data_with_insurance = data[data[polisa_columns].sum(axis=1) != 0]
```

```python
data_with_insurance
```

```python
# data_with_insurance = data[data['RODZAJ_UBEZPIECZENIA'] != 0]
```

```python
# data_with_insurance_drop = data_with_insurance.drop(columns=['RODZAJ_UBEZPIECZENIA'])
data_with_insurance_drop = data_with_insurance.drop(columns=polisa_columns)
```

```python
data_with_insurance_drop
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(data_with_insurance_drop)
data_scaler = scaler.transform(data_with_insurance_drop)
```

```python
data_scaler
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

number_of_clusters = 10

kmeans = KMeans(n_clusters=number_of_clusters, random_state=0).fit(data_scaler)
hierarch = AgglomerativeClustering(n_clusters=number_of_clusters).fit(data_scaler)
```

```python
kmeans.labels_
```

```python
data_with_insurance['k_means'] = kmeans.labels_
data_with_insurance['hierarch'] = hierarch.labels_
```

```python
data_with_insurance
```

```python
for i in range(number_of_clusters):
    var = data_with_insurance[data_with_insurance['k_means'] == i][polisa_columns].var()
#     print(data_with_insurance[data_with_insurance['k_means'] == i])
    print(i, data_with_insurance[data_with_insurance['k_means'] == i]['KWOTA_UZNANIA'].mean())
#     print(i, var)
```

```python
for i in range(number_of_clusters):
    var = data_with_insurance[data_with_insurance['hierarch'] == i][polisa_columns].var()
    print(i, var)
```
